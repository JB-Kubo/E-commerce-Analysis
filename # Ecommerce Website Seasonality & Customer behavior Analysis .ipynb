### RFM Recency, Frequency, Monetary Value of Customers <a id="RFM"></a>

```python
#hypothetical snapshot_day data as if were doing analysis recently
snapshot_date = max(df['InvoiceDate']) + datetime.timedelta(days=1)

# Aggregate data on a customer level
rfm = df.groupby(['CustomerID']).agg({'InvoiceDate': lambda x: (snapshot_date - x.max()).days,\
                                               'InvoiceNo': 'count','Total': 'sum'})
# Rename columns for easier interpretation
rfm.rename(columns = {'InvoiceDate': 'Recency','InvoiceNo': 'Frequency',\
                           'Total': 'MonetaryValue'},inplace=True)

''' Recency - Days since last purchase
    Frequency - count of total invoices
    MonetaryValue - total amout spet 
'''

# Check the first rows
rfm.head()
```

```python
#Recency Frequency and Monetary quartiles

#Recency quartiles - set range() in negative 1,longer days have lower value,set 4 as best and most Recent 
r_labels = range(4, 0, -1) 
r_quartiles = pd.qcut(rfm['Recency'], 4, labels = r_labels)
rfm = rfm.assign(R = r_quartiles.values)

#labling segments from 1-4,  4th segment is the highest value 
labels = range(1,5)

#frequency quartiles 
f_quartiles = pd.qcut(rfm['Frequency'], 4, labels = labels)
rfm = rfm.assign(F = f_quartiles.values)

#Monetray quartiles 
m_quartiles = pd.qcut(rfm['MonetaryValue'], 4, labels = labels)
rfm = rfm.assign(M = m_quartiles.values)

#Check the first rows 
rfm.head()
```

```python
# Build RFM segment with RFM score 
def join_rfm(x): return str(x['R']) + str(x['F']) + str(x['M'])
rfm['RFM_Segment'] = rfm.apply(join_rfm, axis=1) 
rfm['RFM_Segment'] = rfm.apply(join_rfm, axis=1 ) # run twice to insure string clean outcome* 
rfm['RFM_Score'] = rfm[['R','F','M']].sum(axis=1)
rfm.head()
```

```python
#Largest RFM segments
rfm.groupby('RFM_Segment').size().sort_values(ascending=False)[:10]
```

```python
# Customers devided into segments  

'''Use RFM score to group customers into Diamond, Platinum, Gold,Silver rand Bronze segments'''

def segment_me(df): 
    if df['RFM_Score'] >= 11:
        return '1.Diamond'
    elif (df['RFM_Score'] >= 9) and (df['RFM_Score'] < 11) :
        return '2.Platinum'
    elif (df['RFM_Score'] >= 7) and (df['RFM_Score'] < 9) :
        return '3.Gold'
    elif (df['RFM_Score'] >= 5) and (df['RFM_Score'] < 7):
        return '4.Silver'
    else:
        return'5.Bronze'

# create new segment column and apply segmentation on data     
rfm['General_Segment'] = rfm.apply(segment_me, axis=1)

#check the data with the Average values of each column 
rfm.groupby('General_Segment').agg({'Recency': 'mean',\
                                         'Frequency': 'mean','MonetaryValue': ['mean', 'count']}).round(1)
```

### Kmeans <a id="Kmeans"></a>

**Customers Segmentation with Kmeans Clustering** 

```python
''' Kmeans model and Data preparation'''

#checking the distribution of the data, unswek if needed
plt.figure(figsize = [20,5])
plt.subplot(1,3,1)
sns.histplot(rfm['Recency'])
plt.subplot(1,3,2)
sns.histplot(rfm['Frequency'])
plt.subplot(1,3,3)
sns.histplot(rfm['MonetaryValue'])
plt.show();
```

```python
# creatre new data frame only for 'Recency','Frequency','MonetaryValue' to work with Kmeans
data_rfm = rfm[['Recency','Frequency','MonetaryValue']]

# data exponentiation using numpy log   
data_rfm_log = np.log(data_rfm)

# Automatic Scale of the data and fit
scaler = StandardScaler()
scaler.fit(data_rfm_log)

# Scale and center the data
data_normalized = scaler.transform(data_rfm_log)

# Create a pandas DataFrame - store for clustering 
data_normalized = pd.DataFrame(data_normalized, index=data_rfm.index, columns=data_rfm.columns)

# check mean and std if scaled 
print('mean: ', data_normalized.mean(axis=0).round(2))
print('std: ', data_normalized.std(axis=0).round(2))
```

```python
#plot and check Scaled Data if distributed and normalized for Kmeans  
plt.figure(figsize = [20,5])
plt.subplot(1,3,1)
sns.histplot(data_normalized['Recency'])
plt.subplot(1,3,2)
sns.histplot(data_normalized['Frequency'])
plt.subplot(1,3,3)
sns.histplot(data_normalized['MonetaryValue'])
plt.show();
```

**data is ready and scaled to be implemented into Kmeans** 

```python
# Fit KMeans and calculate SSE 'sum of squared errors' for each k
plt.figure(figsize=(10, 5))
sse = {}
for k in range(1, 10):
  
    # Initialize KMeans with k clusters
    kmeans = KMeans(n_clusters=k, random_state=1)
    
    # Fit KMeans on the normalized dataset
    kmeans.fit(data_normalized)
    
    # Assign sum of squared distances to k element of dictionary
    sse[k] = kmeans.inertia_
    plt.title('The Elbow Method')

# Add X-axis label "k"
plt.xlabel('k')

# Add Y-axis label "SSE"
plt.ylabel('SSE')

# Plot SSE values for each key in the dictionary

sns.pointplot(x=list(sse.keys()), y=list(sse.values()))
plt.show()
```

**intuitivly from the Elbow Method plot we can decide on having 4 clusters in Kmeans** 

```python
kmeans = KMeans(n_clusters=4, random_state= 42) #randomstate 42, why? -> check Hitchhikers Guide to the Galaxy
kmeans.fit(data_normalized)
cluster_labels = kmeans.labels_

# Create a DataFrame by adding a new cluster label column
data_rfm_k4 = data_rfm.assign(Cluster=cluster_labels)

# Group the data by cluster
grouped = data_rfm_k4.groupby(['Cluster'])

# Calculate average RFM values and segment sizes per cluster value
grouped.agg({
    'Recency': 'mean',
    'Frequency': 'mean',
    'MonetaryValue': ['mean', 'count']
  }).round(1)
```

### Time Series 

**Monthly and Weekly Percent Change through the whole year per Segment**

```python

# set up the funktion
def monthly_total (data, name): # data = givin data , name = its name 
   
    ''' make subplots of the following:
    1- totaly monthly revenue for givin data 
    2- total revenue pct change each 2 weeks 
    '''
    
    plt.figure(figsize = [20,7])
    plt.subplot(1,2,1)
    sums = data.groupby(['General_Segment', 'month'])['Total'].sum().reset_index(level=0)
    for company in sums['General_Segment'].unique():
        sums[sums['General_Segment'] == company]['Total'].plot();
    plt.xlabel('Month')
    plt.ylabel('Total Income')
    plt.title(name + ' Total Monthly income') 

    plt.subplot(1,2,2)
    data_day = data
    data_day = data_day.set_index('InvoiceDate')
    data_day = data_day.resample(rule = '2W').last()
    returns = data_day['Total'].pct_change()
    returns.plot() 
    plt.title(name + ' pct change each 2 weeks') 
    plt.ylabel('Revenue % change ')
    plt.xlabel('Month')

# plot each segment 
monthly_total(diamond,'Diamond')
monthly_total(platinum,'Platinum')
monthly_total(gold,'Gold')
monthly_total(silver,'Silver')
monthly_total(bronze,'Brozne')
```

**Monthly series shows a clear jump around October, for the christmas seasonality, yet we also see some jump all year round that could be leveraged**

### Heatmaps & Correlations 

**lets check for urchasing behavior correlations for each segment**

### Lemmatized groups <a id="Lemmatized-groups"></a>

```python
# set lemmatize funktion 
def lemmatize (data):
    
    '''lemmatize a givin dataframe, return top 10 tokenized an lemmatized words'''
    
    data = data.dropna()
    wordnet_lemmatizer = WordNetLemmatizer()
    data['tokenized'] = data.apply(lambda row: nltk.word_tokenize(row['Description']), axis=1)
    our_list = data["tokenized"].tolist()
    new_list = []
    for words in our_list:
        new_list += words
    lower_tokens = [t.lower() for t in new_list]
    alpha_only = [t for t in lower_tokens if t.isalpha()]
    english_stops = set(stopwords.words('english'))
    no_stops = [t for t in alpha_only if t not in english_stops]
    lemmatized = [wordnet_lemmatizer.lemmatize(t) for t in no_stops]
    bow = Counter(lemmatized)
    return bow.most_common(10)

# Lemmatizing each segment, and general positive values data 
lem_diamond = lemmatize(diamond)
lem_platinum =lemmatize(platinum)
lem_gold = lemmatize(gold)
lem_silver = lemmatize(silver)
lem_bronze = lemmatize(bronze)
lem_df = lemmatize(df)
```

```python
# defining a bar plotting funktion 
def most_used_words (data, name) : 
    
    ''' plots a barplot from a list of tuples on 1key,1value basis'''
    
    x = [i[0] for i in data]
    y = [i[1] for i in data]
    plt.figure(figsize = (20,3))
    plt.bar (x= x , height= y )
    plt.ylabel('items sold')
    plt.title(name + ' top sold items short description')
    
# plot most used words for all positive data and  per segment
most_used_words(lem_df,'all Customers')
most_used_words(lem_diamond,'Diamond Customers')
most_used_words(lem_platinum,'Platinum Customers')
most_used_words(lem_gold,'Gold Customers')
most_used_words(lem_silver,'Silver Customers')
most_used_words(lem_bronze,'Brozne Customers')
```

### Apriori Associations <a id="Apriori-Associations"></a>

```python

df_grouped= (df 
          .groupby(['InvoiceNo', 'Description'])['Quantity'] 
          .sum().unstack().reset_index().fillna(0) 
          .set_index('InvoiceNo')) 
df_grouped
```

```python
# encode datafram, fill  1 if found 
def hot_encode(x): 
    if x<= 0: 
        return 0
    if x>= 1: 
        return 1
    
# apply 1,0 if descriptiond found or not 
product_sets  = df_grouped.applymap(hot_encode) 
```

```python
#check data if ready 
product_sets.head()
```

```python
# Building the model 
frq_items = apriori(product_sets, min_support = 0.03, use_colnames = True)

# Collecting the inferred rules in a dataframe 
df_asso = association_rules(frq_items, metric ="lift", min_threshold = 0) 
df_asso = df_asso.sort_values(['confidence', 'lift'], ascending =[False, False]) 
```

```python
# check top 10 associations - or above 50% Confidence
df_asso.reset_index(inplace= True, drop =True)
df_asso[df_asso['confidence'] > 0.5]
```

**To construct association rules between elements or items, the algorithm considers 3 important factors which are, support, confidence and lift. Each of these factors is explained as follows:**
- **Support**: the ratio between itemX and the total number of transactions  **itemX / total**

- **Confidence** the proportion of transactions with item X1, in which item X2 also appears by total of item X1 **itemX,item1X2 / X1**

- **Lift** lift is the ratio between the confidence and support : **confidenceX1X2 / Support X2**
